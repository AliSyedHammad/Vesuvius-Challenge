```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

def get_shape_info():
    fragments = {'train': ['1', '2', '3'], 'test': ['a', 'b']}
    data = []

    for set_name, folder_list in fragments.items():
        for folder_name in folder_list:
            file_path = f'/kaggle/input/vesuvius-challenge-ink-detection/{set_name}/{folder_name}/mask.png'
            img = cv2.imread(file_path, cv2.IMREAD_ANYDEPTH)
            if img is not None:  # Ensure the file exists and is read properly
                shape = img.shape
                data.append({
                    'Sample': f'{set_name} {folder_name}',
                    'Shape': shape,
                    'Number of pixels': shape[0] * shape[1]
                })

    # Create DataFrame
    df = pd.DataFrame(data).set_index('Sample')
    return df

# Generate the shape information DataFrame
df = get_shape_info()

# Visualization
plt.figure(figsize=(12, 5))

# Bar plot for pixel sizes
plt.subplot(1, 2, 1)
sns.barplot(x=df.index, y=df['Number of pixels'])
plt.title("Size in pixels for each scan")
plt.ylabel("Number of pixels")
plt.xticks(rotation=45)

# Pie chart for pixel percentages
plt.subplot(1, 2, 2)
plt.pie(df['Number of pixels'], labels=df.index, autopct='%0.1f%%')
plt.title("Pixel size as percentage")

plt.tight_layout()
plt.show()

# Display the DataFrame
df
```

```python
def display_png_files():
    # Train set
    files = ['ir.png', 'inklabels.png', 'mask.png']
    plt.figure(figsize=(12, 18))
    k = 1
    for i in range(1, 4):
        folder = f'/kaggle/input/vesuvius-challenge-ink-detection/train/{i}/'
        for file in files:
            filepath = folder + file
            img = cv2.imread(filepath, cv2.IMREAD_ANYDEPTH)
            plt.subplot(3, 3, k)
            plt.imshow(img, cmap='gray')
            plt.title(f"Papyrus {i} | {file}")
            k += 1
    plt.show()
    
    # Test set
    plt.figure(figsize=(12, 9))
    for k, a in enumerate(['a', 'b']):
        filepath = f'/kaggle/input/vesuvius-challenge-ink-detection/test/{a}/mask.png'
        img = cv2.imread(filepath, cv2.IMREAD_ANYDEPTH)
        img.shape
        plt.subplot(1, 2, k+1)
        plt.imshow(img, cmap='gray')
        plt.title(f"Papyrus {a} | mask.png")
    plt.show()
        
display_png_files()
```

```python
N_SCAN_LAYERS = 65

# Load the volume of the given papyrus fragment
def load_volume(set_name, folder_name):
    path = f'/kaggle/input/vesuvius-challenge-ink-detection/{set_name}/{folder_name}/surface_volume'
    file_format = "{:02d}.tif"
    volume = []
    
    for i in tqdm(range(N_SCAN_LAYERS)):
        file = f'{path}/{file_format.format(i)}'
        img = cv2.imread(file, cv2.IMREAD_ANYDEPTH)
        volume.append(img)
        
    return volume

# Display the volume as a video
def display_volume(set_name, folder_name):
    volume = load_volume(set_name, folder_name)
    fig, ax = plt.subplots()
    camera = Camera(fig)
    
    for i in range(N_SCAN_LAYERS):
        ax.axis('off')
        ax.text(0.5, 1.08, f"Layer {i+1}/{N_SCAN_LAYERS}", fontweight='bold', fontsize=18,
                transform=ax.transAxes, horizontalalignment='center')
        ax.imshow(volume[0], cmap='gray')
        camera.snap()
        del volume[0]
        gc.collect()
        
    plt.close(fig)
    
    animation = camera.animate()
    fix_video_adjust = \
    '<style> video {margin: 0px; padding: 0px; width:100%; height:auto;} </style>'
    display(HTML(fix_video_adjust + animation.to_html5_video()))
    
    del camera
    del animation
    gc.collect()
```

```python
display_volume('train', '1')
```
